# **Imports**

```python
from transformers import pipeline
# from transformers import GPT3Tokenizer, GPT3ForCausalLM

context = """
Thermodynamics is a branch of physics that explores the relationships between heat, work, and energy. It encompasses fundamental laws and
principles governing the behavior of energy in various systems. The first law of thermodynamics, also known as the law of energy
conservation, states that energy is neither created nor destroyed but can only be transferred or transformed from one form to another.
The second law of thermodynamics focuses on the concept of entropy, which measures the level of disorder or randomness in a system.
It states that the entropy of an isolated system tends to increase over time, leading to a decrease in the available energy for useful work.
Another significant principle is the third law of thermodynamics, which indicates that absolute zero temperature cannot be achieved.
Absolute zero represents the lowest possible temperature where the thermal energy of a substance reaches its minimum. Thermodynamics
also encompasses concepts such as heat engines, efficiency, the Carnot cycle, and the Kelvin scale, providing insights into the
conversion of thermal energy into mechanical work and the relationships between temperature and energy. Through mathematical equations
and fundamental principles, thermodynamics plays a crucial role in understanding and analyzing the behavior of energy in various
natural and engineered systems. Thermodynamics plays a vital role in our understanding of the behavior of gases, liquids, and solids.
It provides a framework to analyze and predict how energy transfers occur during physical and chemical processes. The laws of
thermodynamics govern numerous phenomena, ranging from the expansion and compression of gases to the phase transitions of substances.
Engineers and scientists rely on thermodynamics to design efficient engines, power plants, refrigeration systems, and other devices.
The field has practical applications in industries such as aerospace, automotive, energy, and materials science. By studying thermodynamics,
we gain insights into the limits and possibilities of energy conversion, efficiency, and the behavior of complex systems. Furthermore,
thermodynamics has broader implications beyond the physical realm, as its principles can be applied to economic systems, environmental
processes, and even biological systems, aiding our understanding of the natural world and guiding technological advancements.
"""

nlp = pipeline("question-answering")

question1 = "What does the third law of thermodynamics say?"
answer1 = nlp(question=question1, context=context)

question2 = "Why do engineers and scientis rely on thermodynamics?"
answer2 = nlp(question=question2, context=context)

question3 = "What is the First law of thermodynamics?"
answer3 = nlp(question=question3, context=context)
```

# **Query & Results**

```python
print("Context:")
print(context)

print("Query1:", question1)
print("Answer1:", answer1["answer"])

print("Query2:", question2)
print("Answer2:", answer2["answer"])

print("Query3:", question3)
print("Answer3:", answer3["answer"])
```

# Quiz **Generator**

```python
import nltk
nltk.download('punkt')


context = """
Thermodynamics is a branch of physics that explores the relationships between heat, work, and energy. It encompasses fundamental laws and
principles governing the behavior of energy in various systems. The first law of thermodynamics, also known as the law of energy
conservation, states that energy is neither created nor destroyed but can only be transferred or transformed from one form to another.
The second law of thermodynamics focuses on the concept of entropy, which measures the level of disorder or randomness in a system.
It states that the entropy of an isolated system tends to increase over time, leading to a decrease in the available energy for useful work.
Another significant principle is the third law of thermodynamics, which indicates that absolute zero temperature cannot be achieved.
Absolute zero represents the lowest possible temperature where the thermal energy of a substance reaches its minimum. Thermodynamics
also encompasses concepts such as heat engines, efficiency, the Carnot cycle, and the Kelvin scale, providing insights into the
conversion of thermal energy into mechanical work and the relationships between temperature and energy. Through mathematical equations
and fundamental principles, thermodynamics plays a crucial role in understanding and analyzing the behavior of energy in various
natural and engineered systems. Thermodynamics plays a vital role in our understanding of the behavior of gases, liquids, and solids.
It provides a framework to analyze and predict how energy transfers occur during physical and chemical processes. The laws of
thermodynamics govern numerous phenomena, ranging from the expansion and compression of gases to the phase transitions of substances.
Engineers and scientists rely on thermodynamics to design efficient engines, power plants, refrigeration systems, and other devices.
The field has practical applications in industries such as aerospace, automotive, energy, and materials science. By studying thermodynamics,
we gain insights into the limits and possibilities of energy conversion, efficiency, and the behavior of complex systems. Furthermore,
thermodynamics has broader implications beyond the physical realm, as its principles can be applied to economic systems, environmental
processes, and even biological systems, aiding our understanding of the natural world and guiding technological advancements.
"""
# Tokenize the context into sentences
sentences = nltk.sent_tokenize(context)

# Generate questions for each sentence in the context
for sentence in sentences:
    # Tokenize the sentence into words
    words = nltk.word_tokenize(sentence)

    # Remove punctuation marks
    words = [word for word in words if word.isalnum()]

    # Generate a basic question
    if words:
        question = "What is " + words[1] + " " +words[2] + " " +words[3]+"?"
        print("Question:", question)
        print("Answer:", sentence)
        print()
```

```python
import spacy
from transformers import T5ForConditionalGeneration, T5Tokenizer

# Load the T5 model
model_name = "t5-base"
model = T5ForConditionalGeneration.from_pretrained(model_name)

# Load the SpaCy English model
nlp = spacy.load("en_core_web_sm")

# Provide a context or passage
context = "Hugging Face is a company that specializes in Natural Language Processing (NLP) technologies."

# Tokenize the context using SpaCy
doc = nlp(context)
tokens = [token.text for token in doc]

# Join the tokens into a single string
context_tokens = " ".join(tokens)

# Initialize the T5 tokenizer
tokenizer = T5Tokenizer.from_pretrained(model_name)

# Tokenize the context using T5 tokenizer
input_ids = tokenizer.encode(context_tokens, return_tensors="pt")

# Generate questions
question_ids = model.generate(input_ids, max_length=64, num_return_sequences=5, num_beams=4, early_stopping=True)
questions = tokenizer.batch_decode(question_ids, skip_special_tokens=True)

# Print the generated questions
for question in questions:
    print(question)
```
